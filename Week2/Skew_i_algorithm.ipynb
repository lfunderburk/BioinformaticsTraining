```python
import numpy as np
from pathlib2 import Path
import sys
```

```python
def Skew(input_sequence):
    length = len(input_sequence)
    skew = 0
    arr = []
    arr.append(skew)
    for i in range(length):
        if input_sequence[i]=='G':
            skew = skew + 1
        elif input_sequence[i]=='C':
            skew = skew -1
        else:
            skew = skew + 0
        arr.append(skew)
    np_arr = np.array(arr)
    return np_arr
```

```python
def find_min(list_of_numbers):
    save_min = []
    for i in range(len(list_of_numbers)):
        if list_of_numbers[i]==np.min(list_of_numbers):
            save_min.append(i)
    np_save_min = np.array(save_min)
    return np_save_min
```

```python
result = Skew('')
print(result)
minimum = find_min(result)
print(minimum)
```

```python
res = Skew("CATTCCAGTACTTCGATGATGGCGTGAAGA")
res_2 = find_min(res)
print(res)
print(res_2)
```
